```python
import pandas as pd
import numpy as np
```

# Read Dataset

```python
ctg = pd.read_csv('Cardiotocographic_normalisasi.csv')
ctg
```

```python
# Memilih Atribut kolom
x_train=ctg[['LBE','LB','AC','FM','UC','ASTV','MSTV','ALTV',
           'MLTV','DL','DS','DP','DR','AD','DE','LD','FS','SUSP','CLASS']]
x_train
```

# Clustering

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=3)
kmean
```

```python
# Hasil Clustering
y_cluster = kmean.fit_predict(x_train)
y_cluster
```

```python
# Menambahkan kolom hasil Cluster pada dataset
ctg['Cluster']=y_cluster
ctg
```
